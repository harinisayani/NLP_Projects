```python
import pandas as pd
import nltk
nltk.download('stopwords')
import nltk,string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
```

```python
data=pd.read_csv(r'D:\Documents\Harini Projects\AI\file.csv')
data
```

```python
lemmatizer=WordNetLemmatizer()
stop_words=set(stopwords.words('english'))
stop_words
```

```python
def processed(question):
    words=word_tokenize(question)
    processed_words=[]
    for word,tag in pos_tag(words):
        if tag.startswith('NN'):
            pos='n'
        elif tag.startswith('VB'):
            pos='v'
        elif tag.startswith('JJ'):
            pos='a'
        else:
            pos='n'
        if word.lower() not in stop_words and word.lower() not in string.punctuation:
            lemma_word=lemmatizer.lemmatize(word,pos)
            processed_words.append(lemma_word)
    processed_question='-'.join(processed_words)
    return processed_question
```

```python
data['processed_questions']=data['questions'].apply(processed)
data
```

```python
data[['processed_questions','answers']].to_csv(r'D:\Documents\Harini Projects\AI\new.csv',index=False)
output_file=pd.read_csv(r'D:\Documents\Harini Projects\AI\new.csv')
```

```python
def get_answer(question):
    T.delete("1.0",END)
    preprocessed=processed(question)
    matched_rows=output_file[output_file['processed_questions']==preprocessed.strip()]
    if not matched_rows.empty:
        T.insert(END,matched_rows['answers'].iloc[0])
    else:
        T.insert(END,"Sorry.. Can't Find Answer")
```

```python
import speech_recognition as sr
recognizer = sr.Recognizer()
texts=""
def speech_to_text():
    global texts
    texts=""
    with sr.Microphone() as source:
        print("Please say something...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)
        try:
            print("Recognizing...")
            text=recognizer.recognize_google(audio)
            print(f"You said: {text}")
            texts=texts+text
        except sr.UnknownValueError:
            print("Sorry, I couldn't understand what you said. Please try again.")
        except sr.RequestError:
            print("Sorry, I couldn't connect to the speech recognition service.")  
    variable.set(texts)
```

```python
from tkinter import *
root = Tk()
root.geometry("700x400+0+1")
root.config(bg="pink")
Label(root,text="Enter Question: ",font=('arial',20,'bold'),bg="green",fg="white").pack(side=TOP,fill=BOTH)
variable=StringVar()
Entry(root,font=('arial',15,'bold'),bd=7,width=29,textvariable=variable).pack(side=TOP,fill=BOTH,ipady=10)
Button(root,font=('arial',15,'bold'),bd=3,width=29,text="Generate",command=speech_to_text).pack(side=TOP,ipady=4,pady=10)
Button(root,font=('arial',15,'bold'),bd=3,width=29,text="Generate Answer",command=lambda:get_answer(texts)).pack(side=TOP,ipady=4,pady=10)
T=Text(root,font=('arial',15,'bold'),bg="pink",fg="white",bd=4)
T.pack(side=TOP,ipady=4,pady=10)
mainloop()
```
